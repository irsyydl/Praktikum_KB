# IMPORT LIBRARY

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("dataset.csv")
df.head()
```

# Grafik Batang

```python
plt.figure(figsize=(10, 5)) # Digunakan untuk mengatur besar grafik
sns.countplot(data=df, x="dot", palette="flare", hue="country")
plt.title("Kondisi Mesin Es Krim Pada Tiap Negara")
plt.xlabel("Kondisi")
plt.ylabel("Jumlah Mesin")
plt.show()
```

#### Dari data diatas didapatkan kesimpulan bahwa mesin es krim yang paling banyak bekerja berada di USA, kesimpulan lainnya adalah USA memiliki mesin es krim paling banyak dibandingkan negara lainnya

# Grafik Scatter Plot

```python
sns.set_style("darkgrid")
plt.figure(figsize=(10, 5))
plt.title("Grafik Latitude dan Longitude Lokasi Mesin Es Krim Pada Tiap Negara")
plt.xlabel("Latitude")
plt.ylabel("Longitude")
sns.scatterplot(data=df, x="lat", y="lon", hue="country")
plt.show()
```

#### Dari data diatas bisa dilihat penyebaran lokasi mesin es krim mcdonald

# Grafik Pie

```python
data_pie = df["country"].value_counts()
data_pie.head()
```

```python
data_pie = df["country"].value_counts().rename_axis("Country")
data_pie.head()
```

```python
data_pie = df["country"].value_counts().rename_axis("Country").reset_index(name="Id_count")
data_pie.head()
```

```python
plt.title("Persentase Negara Yang Dijadikan Sampel")
plt.pie(data_pie["Id_count"], labels=data_pie["Country"], startangle=90, autopct="%.2f%%",shadow=True)
plt.show()
```
